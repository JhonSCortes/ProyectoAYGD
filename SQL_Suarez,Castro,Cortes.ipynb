```python
user = "root"
password = "root"
host = "localhost"
database = "MAMMO"

config = {
    'user': user,
    'password': password,
    'host':host
}
```

```python
try:
    conn = pymysql.connect(**config)
    cursor = conn.cursor()
    cursor.execute(f"CREATE DATABASE {database};")
    print(f"Se creó la base de datos {database}")

except pymysql.Error as err:
    if err.errno == errorcode.ER_ACCESS_DENIED_CHANGE_USER_ERROR:
        print("Usuario o contraseña incorrecta")
    elif err.errno == errorcode.ER_BAD_DB_ERROR:
        print("La base de datos no existe")
    else:
        print(err)
finally:
    cursor.close()
    conn.close()
```

```python
config = {
    'user': user,
    'password': password,
    'host':host,
    'database': database
}
```

```python
try:
    # Conectar al servidor MySQL usando la configuración especificada
    conn = pymysql.connect(**config)
    cursor = conn.cursor()  # Crear un cursor para ejecutar comandos SQL

    # Crear la tabla si no existe
    create_table_query = """
    CREATE TABLE IF NOT EXISTS resultados (
        Id INT PRIMARY KEY, # Identificador del paciente
        edad INT,                  # Edad del paciente
        BI_RADS INT,                    # Resultado del Paciente
        Ano_Estudio INT,             # año de estudio del paciente
        Edad_buckets DOUBLE           # Puerto de embarque
    );
    """

    # Ejecutar la consulta para crear la tabla
    cursor.execute(create_table_query)  # Ejecuta la consulta SQL
    print("Tabla 'Resultados' creada exitosamente.")  # Mensaje de éxito

except mysql.connector.Error as err:  # Captura cualquier error de MySQL
    if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
        print("Usuario o contraseña incorrectos.")  # Mensaje si hay problemas de acceso
    elif err.errno == errorcode.ER_NO_SUCH_TABLE:
        print("La tabla no existe.")  # Mensaje si la tabla especificada no existe
    else:
        print(err)  # Imprimir otro tipo de error que pueda ocurrir

finally:
    # Cerrar el cursor y la conexión para liberar recursos
    cursor.close()  # Cerrar el cursor
    conn.close()  # Cerrar la conexión
```

```python
csv_file = 'DATASETLIMPIO.csv'

try:
    df = pd.read_csv(csv_file)
    connection_string = f'mysql+pymysql://{user}:{password}@{host}/{database}'
    engine = create_engine(connection_string)
    df.to_sql('resultados', con = engine, if_exists='replace', index=False)

except mysql.connector.Error as err:
    print(f"Error a cargar datos {err}")
finally:
    print("datos cargados")
```

```python
def select(config = config, table = "resultados"):
    conn = pymysql.connect(**config)
    cursor = conn.cursor()
    cursor.execute(f"SELECT * FROM `{table}`")
    resultados = cursor.fetchall()
    for fila in resultados:
        print(fila)
    cursor.close()
    conn.close()
```
